```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

slider = widgets.FloatSlider(value=60, min=0, max=100, step=10, description="Frecuencia f:")
display(slider)

def actualizar_grafica(f):
    t = np.linspace(0, 1, 1000)
    x = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * f * t)
    x2 = np.sin(2 * np.pi * f * t)
    convol = np.convolve(x, x2)

    clear_output(wait=True)
    display(slider)

    plt.figure(figsize=(14, 2))
    plt.subplot(1, 2, 1)
    plt.plot(x, label="x = sin(2π5t) + sin(2πft)")
    plt.title("Señal x(t)")
    # plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(x2, label="x2 = sin(2πft)", color='r')
    plt.title("Señal x2(t)")
    # plt.legend()

    plt.show()

    plt.figure(figsize=(13.78, 2))
    plt.plot(convol, label="Convolución", color='g')
    plt.title("Convolución de x y x2")
    # plt.legend()
    plt.show()

# Conectar el slider con la función de actualización
widgets.interactive(actualizar_grafica, f=slider)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

x = np.sin(np.linspace(0, 20* np.pi, 100)) + np.random.normal(0, 0.1, 100)
x2 = np.cos(np.linspace(0, 20* np.pi, 100)) + np.random.normal(0, 0.1, 100)

wav = 'db4'
a_x, d_x = pywt.dwt(x, wav)
a_x2, d_x2 = pywt.dwt(x2, wav)

convol = np.concatenate((a_x, d_x))

plt.figure(figsize=(16,5))
plt.plot(convol, label='Coeficientes Wavelet')
plt.legend()
plt.show()
```

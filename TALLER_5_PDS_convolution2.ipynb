```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

slider = widgets.FloatSlider(value=60, min=0, max=100, step=10, description="Frecuencia f:")
display(slider)

def actualizar_grafica(f):
    t = np.linspace(0, 1, 1000)
    x = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * f * t)
    x2 = np.sin(2 * np.pi * f * t)
    x2n = np.copy(x2)
    x2n[300:] = 0
    convol = np.convolve(x, x2n)

    clear_output(wait=True)
    display(slider)

    plt.figure(figsize=(14, 2))
    plt.subplot(1, 2, 1)
    plt.plot(x, label="x = sin(2π5t) + sin(2πft)")
    plt.title("Señal x(t)")

    plt.subplot(1, 2, 2)
    plt.plot(x2n, label="x2 = sin(2πft)", color='r')
    plt.title("Señal x2(t)")

    plt.show()

    plt.figure(figsize=(13.78, 2))
    plt.plot(convol, label="Convolución", color='g')
    plt.title("Convolución de x y x2")
    plt.show()

# Conectar el slider con la función de actualización
widgets.interactive(actualizar_grafica, f=slider)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

slider_f = widgets.FloatSlider(value=60, min=0, max=100, step=10, description="Frecuencia")
slider_shift = widgets.IntSlider(value=300, min=0, max=1000, step=10, description="Desplazamiento")

def actualizar_grafica(f, shift):
    t = np.linspace(0, 1, 1000)
    x = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * f * t)
    x2 = np.sin(2 * np.pi * f * t)

    x2n = np.zeros_like(x2)
    if shift < len(x2):
        x2n[shift:shift+250] = x2[shift:shift+250]

    convol = np.convolve(x, x2n, mode='full')
    convol = convol[:2000]

    clear_output(wait=True)
    display(slider_f, slider_shift)

    plt.figure(figsize=(14, 2))
    plt.subplot(1, 2, 1)
    plt.plot(x, label="x = sin(2π5t) + sin(2πft)")
    plt.title("Señal x(t)")

    plt.subplot(1, 2, 2)
    plt.plot(x2n, label="x2", color='r')
    plt.title("Señal x2(t)")
    plt.show()

    plt.figure(figsize=(13.78, 2))
    plt.plot(convol, label="Convolución", color='g')
    plt.title("Convolución de x y x2")
    plt.show()

widgets.interactive(actualizar_grafica, f=slider_f, shift=slider_shift)
```

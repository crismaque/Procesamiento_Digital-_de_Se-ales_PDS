1. Generar señal con sample rate >> (grande).
2. Generar señal con sample rate << (pequeño).
3. Las dos señales en un intervalo de tiempo (0,1).
4. Calcular FFT para ambas señales, gráficar y comparar.

```python
import numpy as np
import matplotlib.pyplot as plt

T = 1

t1 = np.linspace(0, T, int(1000 * T), endpoint=False)
t2 = np.linspace(0, T, int(5000 * T), endpoint=False)

signal1 = np.sin(60 * 2 * np.pi * t1)
signal2 = np.sin(60 * 2 * np.pi * t2)

plt.figure(figsize=(16, 4))
plt.subplot(1,2,1)
plt.plot(t1, signal1)
plt.title("1000 samples")
plt.grid()
plt.xlim(0,0.5)

plt.subplot(1,2,2)
plt.plot(t2, signal2)
plt.title("5000 samples")
plt.grid()
plt.xlim(0,0.5)

plt.show()
```

```python
Xw1 = np.fft.fft(signal1)
freq1 = np.fft.fftfreq(len(t1), 1/1000)

Xw2 = np.fft.fft(signal2)
freq2 = np.fft.fftfreq(len(t2), 1/5000)

plt.figure(figsize=(16, 6))
plt.plot(freq1[:len(freq1)//2], np.abs(Xw1[:len(Xw1)//2]), label='1000 samples')
plt.plot(freq2[:len(freq2)//2], np.abs(Xw2[:len(Xw2)//2]), label='5000 samples')
plt.title('Espectro')
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.grid()
plt.xlim(0,100)
plt.legend()

plt.show()
```

```python
#PSD
psd1=(np.abs(Xw1)**2)
psd2=(np.abs(Xw2)**2)
```

```python
plt.figure(figsize=(16, 6))
plt.semilogy(freq1[:len(freq1)//2], np.abs(psd1[:len(psd1)//2]), label='1000 samples')
plt.semilogy(freq2[:len(freq2)//2], np.abs(psd2[:len(psd2)//2]), label='5000 samples')
plt.title('Espectro')
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.grid()
plt.xlim(0,100)
plt.legend()

plt.show()
```
